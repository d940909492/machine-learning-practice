```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
data = pd.read_csv("bank.csv",delimiter=';' )
x = data[['age', 'balance']].values
y = np.where(data['housing']=='yes', 0, 1)
```

```python
clf = SVC(kernel='rbf',C = 100)
clf.fit(x,y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)

ax = plt.gca()

x2 = np.linspace(0, 5, 30)
y2 = np.linspace(-10, 0, 30)

x3, y3 = np.meshgrid(x2, y2)
xy = np.vstack([x3.ravel(), y3.ravel()]).T
z = clf.decision_function(xy).reshape(x3.shape)

plt.contour(x2, y2, z, levels=[-1, 0, 1], alpha=0.5,linestyles=['--', '-', '--'])
plt.title('svm testing')
plt.show()
```

## since I cant find any good testing data from web, I use samples generator right now

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples=50, centers=2,random_state=30, cluster_std=0.60)
plt.scatter(x[:, 0], x[:, 1], c = y)
```

```python
clf = SVC(kernel='linear',C=1)
clf.fit(x,y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)

ax = plt.gca()

x2 = np.linspace(0, 5, 30)
y2 = np.linspace(-10, 0, 30)

x3, y3 = np.meshgrid(x2, y2)
xy = np.vstack([x3.ravel(), y3.ravel()]).T
z = clf.decision_function(xy).reshape(x3.shape)

plt.contour(x2, y2, z, levels=[-1, 0, 1], alpha=0.5,linestyles=['--', '-', '--'])
plt.title('SVM Testing')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
